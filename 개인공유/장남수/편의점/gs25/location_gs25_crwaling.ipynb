```python
from selenium import webdriver
```

```python
browser = webdriver.Chrome('c:/informs/chromedriver.exe')
```

```python
url = 'http://gs25.gsretail.com/gscvs/ko/store-services/locations#;'
browser.get(url)
```

```python
def get_store(browser):
    store_list = []
    stores = browser.find_elements_by_css_selector('#storeInfoList > tr')
    for store in stores:
        try:
            name = store.find_element_by_css_selector('a.st_name').text
            add = store.find_element_by_css_selector('a.st_address').text
            loc1 = add.split(' ')[0]
            loc2 = add.split(' ')[1]
            location = [loc1, loc2, name, add]
        except: location = []
        store_list.append(location)
    
    return store_list
```

```python
store_list = get_store(browser)
print(store_list)
```

```python
def next_page(browser):
    button = browser.find_element_by_css_selector('a.next')
    button.click()
```

```python
next_page(browser)
```

```python
import time
```

```python
MAX_PAGE = 2466
# store_list = []
store_list += get_store(browser)
for i in range(MAX_PAGE - 2338):
    next_page(browser)
    browser.implicitly_wait(3)
    time.sleep(1)
    store_list += get_store(browser)
```
