```python
from selenium import webdriver
import time
from bs4 import BeautifulSoup
import pandas as pd
```

```python
browser = webdriver.Chrome('c:/informs/chromedriver.exe')
```

```python
#url 접속
url = 'http://cu.bgfretail.com/store/list.do?category=store&sf=N'
browser.get(url)
```

```python
def get_stores(browser):
    # 화면에 표시된 5개 매장 정보를 리턴한다
    
    html = browser.page_source
    soup = BeautifulSoup(html, 'html.parser')
    result = []
    stores_inpage = soup.select('div.detail_store tbody > tr')

    for store in stores_inpage:
        name = store.select('span.name')[0].text
        try:
            add = store.select('div.detail_info > address > a')[0].text
            loc1 = add.split(' ')[0]
            loc2 = add.split(' ')[1]
        except:
            add = loc1 = loc2 = ''
        result.append([loc1,loc2,name,add])
    return result
```

```python
def move_next(browser):
    
    # 현재 페이지를 찾는다
    page_index = browser.find_elements_by_css_selector('ul.pagination > a')
    for i in range(len(page_index)):
        if page_index[i].get_attribute('class')=='Current':
            page_now = page_index[i].text
            print(' 페이지 조회: {} '.format(int(page_now)+1))
            next_page = page_index[i+1]
            next_page.click()
            browser.implicitly_wait(20)
            time.sleep(1)
            break
        
```

```python
def move_next_5pages(browser):
    browser.find_elements_by_css_selector('ul.pagination > a.Image')[2].click()
```

```python
# 페이지 빨리 넘기기. 

click_count = 1218//5

for _ in range(click_count):
    move_next_5pages(browser)
    time.sleep(1)
```

```python
Max_Page = 2524
# Max_Page = 3
page_now = 1
stores = []

try:
    while page_now <= Max_Page:
        stores += get_stores(browser)
        move_next(browser)
        page_now +=1
except:
    data = pd.DataFrame(stores)
    data.columns = ['loc1','loc2','name','location']
    # data.to_csv('save222.csv', encoding='cp949')
#     data.to_csv('save521.csv', encoding='cp949')
    data.to_csv('save2057.csv', encoding='cp949')
    
data = pd.DataFrame(stores)
data.columns = ['loc1','loc2','name','location']
# data.to_csv('save222.csv', encoding='cp949')
#     data.to_csv('save521.csv', encoding='cp949')
data.to_csv('save2057.csv', encoding='cp949')
```

```python
def simple_loc(x):
    if str(x[-2]) == '북' or str(x[-2]) == '남': return x[0]+x[-2]
    else: return x[:2]
```

```python
datas['loc1'] = datas.loc1.apply(simple_loc)
```

```python
# 세종시는 그냥 세종으로 통합

def change(x):
    if x['loc1'] == '세종': return x['loc1']
    else: return x['loc1'] + ' ' + x['loc2']

datas['loc'] = datas.apply(change, axis = 1)
```
