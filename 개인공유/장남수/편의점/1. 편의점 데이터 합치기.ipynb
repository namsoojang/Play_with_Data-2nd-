# 데이터 살펴보기

```python
stores = pd.read_csv('stores.csv', encoding='cp949', index_col=0)
```

```python
stores.head()
```

```python
from matplotlib import pyplot as plt
from matplotlib import rcParams
import seaborn as sns
rcParams['font.size'] = 12
```

```python
plt.figure(figsize=(4, 3))
stores.sum(axis=0).iloc[:3].plot(kind='bar')
```

```python
import scipy.stats
import numpy as np
```

```python
fig = plt.figure(figsize=(9, 3))

def plot_nstores(s1, s2, label1, label2):
    plt.scatter(stores[s1] + np.random.random(len(stores)),
                stores[s2] + np.random.random(len(stores)),
                edgecolor='none', alpha=0.75, s=6, c='black')
    plt.xlim(-1, 15)
    plt.ylim(-1, 15)
    plt.xlabel(label1)
    plt.ylabel(label2)
    
    r = scipy.stats.pearsonr(stores[s1], stores[s2])
    plt.annotate('r={:.3f}'.format(r[0]), (10, 12.5))

ax = fig.add_subplot(1, 3, 1)
plot_nstores('gs', 'cu', 'GS25', "CU")

ax = fig.add_subplot(1, 3, 2)
plot_nstores('cu', 'seven', 'CU', 'SevenEleven')

ax = fig.add_subplot(1, 3, 3)
plot_nstores('seven', 'gs', "SevenEleven", 'GS25')

plt.tight_layout()
```

```python
plt.figure(figsize=(4, 3))
for col, label in [('cu', 'CU'), ('gs', 'GS25'), ('seven', "SevenEleven")]:
    cumulv = np.cumsum(sorted(stores[col], reverse=True)) / stores[col].sum()
    plt.plot(cumulv, label='{} ({})'.format(label, int(stores[col].sum())))
plt.legend(loc='best')
plt.xlabel('Number of districts (si/gun/gu)')
plt.ylabel('Cumulative fraction')
```

# 인구통계 자료 합치기

```python
kosis = './kosis/kosis.csv'
data = pd.read_csv(kosis, encoding = 'cp949')
```

```python
data.head()
```

```python
data['loc'] = data.apply(lambda x: x['loc1'] + ' ' + x['loc2'], axis = 1)
```

```python
data.head()
```

```python
data.set_index('loc')
```

```python
stores.info()
data.info()
```

```python
data_total = pd.merge(stores, data, how='outer', left_index = True, right_on='loc')
```

```python
data_total.head(10)
```

```python
data_total.columns
```

```python
data_total.columns = ['gs', 'cu', 'seven', 'stores', 'loc1', 'loc2', 'ppl', '1p', '2p',
       '3p', '4p', '5p', '6p', '7pplus', 'loc']
```

```python
data_total.head()
```

```python
data_total['1p_ratio'] = data_total['1p'] / data_total['ppl']
```

```python
data_total.head()
```

```python
data_total.fillna(0)
```

```python
data_total.dropna()
```

```python
data_total.head()
```

```python
data_anal = data_total[['loc','gs','cu','seven','stores','ppl','1p','1p_ratio', 'loc1','loc2']]
data_anal.head()
```

```python
# data_anal['stores'].sum
data_anal.sum(axis=0).loc['stores']
```

```python
stores_sum = data_anal.sum(axis=0).loc['stores']
data_anal['store_ratio'] = data_anal['stores'].apply(lambda x: x/ stores_sum)
```

```python
data_anal.head()
```

```python
data_anal = data_anal.dropna()
```

```python
data_anal.to_csv('stores_anal.csv',encoding='cp949')
```

```python
fig = plt.figure(figsize=(9, 3))

def plot_nstores(stores, s1, s2, label1, label2):
    plt.scatter(stores[s1],stores[s2],
                edgecolor='none', alpha=0.75, s=6, c='black')
#     plt.xlim(-1, 15)
#     plt.ylim(-1, 15)
    plt.xlabel(label1)
    plt.ylabel(label2)
    r = scipy.stats.pearsonr(stores[s1], stores[s2])
#     plt.annotate('r={:.3f}'.format(r[0]), (10, 12.5))
    plt.annotate('r={:.3f}'.format(r[0]), (0.5,0.5))

ax = fig.add_subplot(1, 3, 1)
plot_nstores(data_anal, 'stores', '1p_ratio', 'stores', "1p_ratio")

ax = fig.add_subplot(1, 3, 2)
plot_nstores(data_anal, '1p_ratio', 'store_ratio', '1p_ratio', 'store_ratio')

ax = fig.add_subplot(1, 3, 3)
plot_nstores(data_anal, '1p', 'stores', '1p', 'stores')


# plt.tight_layout()
```
