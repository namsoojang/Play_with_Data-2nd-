```python
from selenium import webdriver
```

```python
import time
```

```python
browser = webdriver.Chrome('c:/informs/chromedriver.exe')
```

```python
# 세븐일레븐 점포 위치 찾기 페이지에 접속한다
def se_location_page(browser):
    url = 'http://www.7-eleven.co.kr/'
    browser.get(url)
    browser.implicitly_wait(3)
    time.sleep(1)
    find_location = browser.find_element_by_css_selector('a.util_store.store_open')
    find_location.click()
```

```python
se_location_page(browser)
```

```python
location_options = browser.find_elements_by_css_selector('#storeLaySido > option')
for i, option in enumerate(location_options):
    print(i, option.text)
```

```python
# 총 17개의 지역으로 나뉘어져 있음  1부터 17까지 입력
def set_location(browser, i):
    location_option = browser.find_elements_by_css_selector('#storeLaySido > option')[i]
    print('지역설정: {}'.format(location_option.text))
    location_option.click()
    #조회버튼 클릭
    browser.find_element_by_css_selector('#storeButton1').click()
    browser.implicitly_wait(3)
    time.sleep(5)
```

```python
set_location(browser, 1)
```

```python
from bs4 import BeautifulSoup
```

```python
# 해당지역의 점포 정포를 가져온다(BeautifulSoup 활용)
def get_stores(browser):
    stores = []
    html = browser.page_source
    soup = BeautifulSoup(html, 'html.parser')
    store_list = soup.select('div.cont_store > div.list_stroe > ul > li > a')
    for store in store_list:
        name = store.select('span')[0].text.strip()
        location = store.select('span')[1].text.strip()
#         loc1 = location.split(' ')[0]
#         loc2 = location.split(' ')[1]
#         store = [loc1, loc2, name, location]
        store = [name, location]
#         print(store)
        stores.append(store)
    return stores
```

```python
stores = get_stores(browser)
```

```python
stores = []
for i in range(1,18):
    set_location(browser, i)
    stores += get_stores(browser)
    print(len(stores))
```

```python
# 띄어쓰기 조정
def loc_mod(s):
    check_list = ['전북','대구시','대구광역시','대구','광주','경남','경기도']
    for check in check_list:
        if check in s and s[len(check)] != ' ':
            s = s.replace(check, check+' ')
            break
#             print(s)
    return s
```

```python
# 띄어쓰기 조정
def loc_mod2(s):
    check_list = ['창원시','화성시','장군면','연기면','연동면','연서면','안산시','시흥시','순천시','사상구','부안군',
                  '보람동','반곡동','동구','달성군','달서구','도담동','남동구','김포시','강서구','고양시','군산시', '삼척시']
    for check in check_list:
        if check in s and s[s.find(check)+len(check)] != ' ':
            s = s.replace(check, check+' ')
#             break
#             print(s)
    return s
```
